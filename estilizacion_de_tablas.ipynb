## Misión 1: Equipo de Ventas

Necesitamos construir una visualización para indicar los **10 principales clientes por total de ventas**.

Con ello, el equipo de ventas puede pensar en estrategias para mantener a estos clientes comprando en nuestra tienda.

```python
clientes=ventas.groupby(['nombre_cliente'])['ventas'].sum().nlargest(10).copy()
clientes=clientes.reset_index()
clientes
```

```python
#Otra forma de renombrar las columnas
#dic={'nombre_cliente':'Clientes','ventas':'Ventas'}
#clientes=clientes.rename(columns=dic)

clientes.columns=['Clientes','Ventas']
clientes['Ranking']=clientes.index+1
clientes.set_index('Ranking',inplace=True)
clientes
```

Para aplicar estilos a un DataFrame, primero debemos convertirlo en un objeto Style


En resumen:

clientes: Muestra el DataFrame sin guardar la visualización.

estilo = clientes.style: Crea un objeto Style que almacena la configuración de estilo, permitiéndote aplicar cambios y visualizarlos cuando lo desees.

```python
estilo=clientes.style
estilo
```

```python
estilo.format({'Ventas':'{:,.2f}'})
estilo
```

## Misión 4: Equipo de Logística

Nuevamente el equipo de logística desea replantear su estructura y organización de provisionamiento de productos e, para ello, se pensó en una visualización que permita identificar el estándar de la **cantidad de productos solicitados por mes por departamento**.

```python
meses=ventas.copy()
meses.sort_values('fecha_pedido', inplace=True)
meses
```

```python
meses['mes']=meses['fecha_pedido'].dt.strftime('%Y - %b')
meses=meses.reset_index(drop=False)
meses.head()
```

```python
pd.set_option('display.max_columns',None)
```

```python
ventas_mensuales=meses.pivot_table(index=meses.departamento,columns='mes',values='cantidad',aggfunc='sum', sort=True)
ventas_mensuales.index.rename('Departamento',inplace=True)

ventas_mensuales
```

```python
estilo_mensual=ventas_mensuales.style
estilo_mensual.set_sticky(axis='index')
estilo_mensual
```

```python
columna_fija_header = {
    'selector': 'thead tr th:nth-child(1)',
    'props': 'position:sticky; left:0px; z-index:3; background-color:#F0F0F0'
}

columna_fija_body = {
    'selector': 'tbody tr th:nth-child(1)',
    'props': 'position:sticky; left:0px; z-index:1; background-color:#F0F0F0; text-transform:capitalize'
}

columnas={
    'selector':'.col_heading',
    'props':'font-weight:normal; font-family:Arial'
    }
tabla={
    'selector':'td,th',
    'props':'text_align:left'
}
indices={
    'selector':'index_name',
    'props':'font-weight:bold; font-family:Arial; text_align:right'
}
estilo_mensual.set_table_styles([columna_fija_header,columna_fija_body,columnas,tabla,indices], overwrite=False)
```

#Exportando una visualizacion
Vamos a transformar el DataFrame que lista los géneros de películas mejor valorados en:

un archivo HTML;

un archivo Excel;

y un archivo LaTeX.

```python
import pandas as pd

# DataFrame con películas bien valoradas
datos = {
    'Género': ['Drama', 'Comedia', 'Acción', 'Ciencia Ficción', 'Terror'],
    'Nota Media': [4.5, 4.3, 4.2, 4.1, 3.8],
    'Cantidad de Evaluaciones': [1500, 1200, 1000, 800, 600]
}

df_peliculas = pd.DataFrame(datos)

# Añade un ranking de popularidad basado en el número de evaluaciones
df_peliculas['Ranking'] = df_peliculas['Cantidad de Evaluaciones'].rank(method='dense', ascending=False).astype(int)

# Define la columna 'Ranking' como índice del DataFrame
df_peliculas.set_index('Ranking', inplace=True)

# Objeto style
s_pelicula = df_peliculas.style

# Crear un archivo HTML
s_pelicula.to_html('estilo_html.html', index=False)

# Crear un archivo Excel
s_pelicula.to_excel('estilo_xlsx.xlsx', engine='openpyxl')

# Crear un archivo LaTeX
with open('estilo_latex.tex', 'w') as f:
    f.write(s_pelicula.to_latex())
```

#Desafio

En el archivo, tenemos un informe de costos de venta de libros y películas de una tienda.
```python
import pandas as pd
tienda=pd.read_csv('https://gist.githubusercontent.com/ahcamachod/52da15c2bbb31e4270b5c267dd051473/raw/62c366878ebd4be2ec899836413c8d1f3e116c01/tienda_libros.csv',sep=',')
tienda.head()
```

```python
tienda['fecha_pedido'] = pd.to_datetime(tienda['fecha_pedido'], format='%Y-%m-%d')
tienda['fecha_llegada'] = pd.to_datetime(tienda['fecha_llegada'], format='%Y-%m-%d')
tienda.info()
```

##El primer desafío es construir una visualización que permita identificar qué países han solicitado más productos por pedidos, para así poder realizar un estudio sobre la distribución y logística de productos.



```python
paises=tienda.groupby(['pais'])['unidades'].sum().nlargest(10).copy()
paises=paises.reset_index()
paises.columns=['Pais','Unidades Pedidas']
paises['Ranking']=paises.index+1
paises.set_index('Ranking',inplace=True)
paises
```

```python
s_paises=paises.style
s_paises
```

##El objetivo de este desafío es construir una visualización que permita identificar qué categorías de productos generaron más y menos costos para la tienda.



```python
costo_producto=tienda.groupby(['categoria'])['costo_producto'].sum()
costo_producto.index.name='Categoria del Producto'
costo_producto.name='Costo del Pedido'
costo_producto
```

```python
s_costo_producto=costo_producto.to_frame().style
s_costo_producto.format('USD {:,.2f}').highlight_max(color='pink').highlight_min(color='lightgreen')

encabezado={
    'selector':'th',
    'props':'font-weight:bold; font-family:Arial; text-align:center'
}
s_costo_producto.set_table_styles([encabezado],overwrite=False)

s_costo_producto
```

##El desafío consiste en construir una visualización que permita informar la cantidad y distribución de pedidos por tipo de descuento. Esto permitirá que los departamentos de la empresa comprendan la demanda de productos en cada promoción.



```python
tienda.head(2)
```

```python
demanda_promocion=pd.DataFrame(tienda['tipo_descuento'].value_counts())
demanda_promocion.columns=['Nº Pedidos']
demanda_promocion.index.name='Tipo de Descuento'

porcentaje=demanda_promocion['Nº Pedidos'].to_numpy()
porcentaje =100*porcentaje/porcentaje.sum()
demanda_promocion['Distribucion de Pedidos']=porcentaje

encabezado={
    'selector':'th',
    'props':'font-weight:bold; font-family:Arial; text-align:left;background-color:white'
}
celdas={
    'selector':'td',
    'props':'text-align: center; background-color:white'
}
estilo_demanda = demanda_promocion.style.set_table_styles([encabezado,celdas])
estilo_demanda.format({'Distribucion de Pedidos':'{:.2f} %'})\
.bar(subset='Distribucion de Pedidos',vmin=0,vmax=100,color='#9cd4bd',height=50,width=50)
estilo_demanda
```

##Nuestro desafío ahora es proporcionar una visualización que permita mostrar el tiempo medio de entrega de pedidos durante los meses de los años 2013 a 2015 por cada país. Esto permitirá analizar el rendimiento del departamento de transporte de la empresa durante esos años y trazar un plan de mejoras.



```python
tienda.head(2)
```

```python
fecha = tienda.copy()

# Filtrar 2013-2015
rango = (fecha['fecha_pedido'] >= '2013-01-01') & (fecha['fecha_pedido'] <= '2015-12-31')
fecha = fecha[rango]

# Crear columnas
fecha['Mes'] = fecha['fecha_pedido'].dt.strftime('%Y - %b')
fecha['demora_entrega'] = (fecha['fecha_llegada'] - fecha['fecha_pedido']).dt.days

# Pivot table: tiempo medio por país y mes
entregas = fecha.pivot_table(
    index='pais',
    columns='Mes',
    values='demora_entrega',
    aggfunc='mean',
    sort=False
)

# Estilo
s_entregas = entregas.style.format('{:,.2f}')
s_entregas.set_sticky(axis="index")
s_entregas.set_table_styles([
    {
        'selector': 'thead tr th:nth-child(1)',
        'props': 'position:sticky; left:0px; z-index:3; background-color:#F0F0F0'
    },
    {
        'selector': 'tbody tr th:nth-child(1)',
        'props': 'position:sticky; left:0px; z-index:1; background-color:#F0F0F0'
    }
])
s_entregas
```

##Nuestro desafío implica una visualización que permita separar los gastos de envío de cada producto (costo total) por tipo de cliente, de modo que sea posible comprender el tipo de producto que generó más costos de envío para cada tipo de cliente. Además, también es interesante mostrar la distribución de estos costos por país enviado, lo que permite entender también los gastos por país.

```python
tienda.head(2)
```

```python
tabla_cliente = tienda.pivot_table(
    index='categoria',
    columns=['tipo_cliente', 'pais'],
    values='costo_envio',
    aggfunc='sum',
    margins=True,
    margins_name='Total'
)

tabla_cliente = tabla_cliente.rename_axis('Categoría', axis=0)
tabla_cliente = tabla_cliente.rename_axis(['Tipo de Cliente', 'País'], axis=1)

n_paises = tienda['pais'].nunique()

def resaltar_celdas(df):
    estilos = pd.DataFrame('background-color: white', index=df.index, columns=df.columns)

    estilos.loc['Total', ('B2B', 'Perú')] = 'background-color: #D8D8D8'
    estilos.loc['Total', ('B2C', 'Paraguay')] = 'background-color: #D8D8D8'
    estilos.loc['Libro', ('Total', '')] = 'background-color: #D8D8D8'

    return estilos

estilo_cliente = tabla_cliente.style.format('{:,.2f}')

estilo_cliente.apply(lambda x: resaltar_celdas(tabla_cliente), axis=None)

estilo_cliente.set_table_styles([
    {
        'selector': f'td:nth-child({n_paises + 2}), th:nth-child({n_paises + 2})',
        'props': 'border-left: 2px solid black'
    },
    {
        'selector': 'td:last-child, th:last-child',
        'props': 'border-left: 2px solid black'
    },
    {
        'selector':'td,th:not(.index_name)',
        'props':'font-weight:normal; font-family:Arial;text-align:center'
    },
    {
        'selector':'.index_name',
        'props':'font-weight:normal; font-style:italic;text-align:right;color:#696969'
    },
    {
        'selector': 'tbody tr:last-child',
        'props': 'border-top: 2px solid black'
    }
], overwrite=False)

estilo_cliente
```

```python
columna_fija_header = {
    'selector': 'thead tr th:nth-child(1)',
    'props': 'position:sticky; left:0px; z-index:3; background-color:white'
}

columna_fija_body = {
    'selector': 'tbody tr th:nth-child(1)',
    'props': 'position:sticky; left:0px; z-index:1; background-color:white; text-transform:capitalize'
}
estilo_cliente.set_table_styles([columna_fija_header, columna_fija_body], overwrite=False)
```
